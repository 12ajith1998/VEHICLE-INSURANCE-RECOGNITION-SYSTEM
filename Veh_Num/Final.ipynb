```python
import csv
```

```python
with open('plate_number.csv', 'rt', encoding='utf-8') as master:
    master_indices = dict((r[0], i) for i, r in enumerate(csv.reader(master)))

    with open('Vehicle Insurance System.csv', 'rt', encoding='utf-8') as hosts:
        with open('result.csv', 'w') as results:    
            reader = csv.reader(hosts)
            writer = csv.writer(results)

            writer.writerow(next(reader, []) + ['RESULTS'])

            for row in reader:
                index = master_indices.get(row[0])
                if index is not None:
                    message = 'Insuranced'
                    writer.writerow(row + [message])

                else:
                     message = 'NOT Insuranced'
                     writer.writerow(row + [message])

        results.close()
```
